```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
df=pd.read_csv(r"C:\Users\ankit\Downloads\diabetes.csv")
print(df.head())
```

```python
df.isnull().mean()
```

```python
print(df.isnull().sum())
```

```python
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0,np.nan)
print(df.isnull().sum())
```

```python
mean1 = df.Glucose.mean()
mean2 = df.BloodPressure.mean()
mean3 = df.SkinThickness.mean()
mean4 = df.Insulin .mean()
mean5 = df.BMI.mean()

print(mean1)
print(mean2)
print(mean3)
print(mean4)
print(mean5)
```

```python
median1 = df.Glucose.median()
median2 = df.BloodPressure.median()
median3 = df.SkinThickness.median()
median4 = df.Insulin.median()
median5 = df.BMI.median()

print(median1)
print(median2)
print(median3)
print(median4)
print(median5)
```

```python
df1 = df[["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]]
df1.head(10)
```

```python
# Histogram
df1.hist(bins=50, figsize=(20, 15))
plt.show()

```

```python
df1["Insulin_Mean"] = df1.Insulin.fillna(mean4)
df1["Insulin_Median"] = df1.Insulin.fillna(median4)
df1.head(20)
```

```python
df1.plot()
```

```python
df1.Insulin.plot()
```

```python
plt.rcParams["figure.figsize"] = [8,6]
fig=plt.figure()
ax = fig.add_subplot(111)
df1.Insulin.plot(kind='kde', ax=ax )
df1.Insulin_Mean.plot(kind='kde', ax=ax, color ='green' )
df1.Insulin_Median.plot(kind='kde', ax=ax, color ='red')
lines,labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

```python
df1.Insulin.hist()
```

```python
df1.Insulin.hist(bins=50)
```

```python
eod_value4 = df1.Insulin.mean() + 3 * df1.Insulin.std() # std()-- standard deviation
print(eod_value4)
```

```python
df1['insulin_eod'] = df1.Insulin.fillna(eod_value4)
df1.head(20)
```

```python
plt.rcParams["figure.figsize"] = [8,6]
fig = plt.figure()
ax = fig.add_subplot(111)
df1['Insulin'] .plot(kind='kde', ax=ax)
df1['insulin_eod'] .plot(kind='kde', ax=ax)
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```
