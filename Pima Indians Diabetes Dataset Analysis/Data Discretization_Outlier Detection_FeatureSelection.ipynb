```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import numpy as np 
plt.rcParams["figure.figsize"] = [8,6] 
sns.set_style("darkgrid") 
df=pd.read_csv(r"C:\Users\ankit\Downloads\diabetes.csv") 
print(df.head()) 
```

```python
df
```

```python
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0,np.nan)
df
```

```python
df1 = df[["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]]
df1.head(20)
```

```python
# Equal Width Discretization
```

```python
sns.histplot(df1['Insulin'])
```

```python
insulin_range=df1['Insulin'].max()-df1['Insulin'].min()
print(insulin_range)
```

```python
insulin_range/10
```

```python
lower_interval = int(np.floor( df1['Insulin'].min())) 
upper_interval = int(np.ceil( df1['Insulin'].max())) 
interval_length = int(np.round(insulin_range / 10)) 
print(lower_interval) 
print(upper_interval) 
print(interval_length) 
```

```python
total_bins = [i for i in range(lower_interval, upper_interval+interval_length, interval_length)] 
print(total_bins) 
```

```python
bin_labels = ['Bin_no_' + str(i) for i in range(1, len(total_bins))] 
print(bin_labels) 
```

```python
df1['insulin_bins'] = pd.cut(x=df1['Insulin'],bins=total_bins, labels=bin_labels, include_lowest=True) 
df1.head(10)
```

```python
df1.groupby('insulin_bins')['Insulin'].count().plot.bar() 
plt.xticks(rotation=45) 
```

```python
# Equal Frequency Discretization
```

```python
discretised_insulin, bins = pd.qcut(df['Insulin'],10, labels=None, retbins=True, precision=3, duplicates='raise') 
pd.concat([discretised_insulin, df['Insulin']], axis=1).head(10) 
```

```python
print(bins)
print(type(bins))
```

```python
bin_labels = ['Bin_no_' +str(i) for i in range(1,11)] 
print(bin_labels) 
```

```python
df1['insulin_bins'] = pd.qcut(x=df1['Insulin'],q=10,labels=bin_labels,duplicates='drop') 
df1.head(20) 
```

```python
df1.groupby('insulin_bins')['Insulin'].count().plot.bar() 
plt.xticks (rotation=45) 
```
